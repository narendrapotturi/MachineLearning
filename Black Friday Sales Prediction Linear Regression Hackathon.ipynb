```python
df['Product_Category_2'].fillna(df['Product_Category_2'].median(), inplace = True)
```

```python
def map_gender(gender):
    if gender == 'M':
        return 1
    else:
        return 0
df['Gender'] = df['Gender'].apply(map_gender)
```

```python
def map_age(age):
    if age == '0-17':
        return 0
    elif age == '18-25':
        return 1
    elif age == '26-35':
        return 2
    elif age == '36-45':
        return 3
    elif age == '46-50':
        return 4
    elif age == '51-55':
        return 5
    else:
        return 6
df['Age'] = df['Age'].apply(map_age)
```

```python
def map_city_categories(city_category):
    if city_category == 'A':
        return 2
    elif city_category == 'B':
        return 1
    else:
        return 0
df['City_Category'] = df['City_Category'].apply(map_city_categories)
```

```python
def map_stay(stay):
        if stay == '4+':
            return 4
        else:
            return int(stay)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].apply(map_stay)   
```

```python
cols = ['Product_Category_3']
df.drop(cols, inplace = True, axis =1)
```

```python
df.head()
```

```python
lr = LinearRegression(normalize = True)
```

```python
x = df.drop(['Purchase'], axis=1)
y = df['Purchase']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56)
```

```python
lr.fit(train_x, train_y)
```

```python
train_predict = lr.predict(train_x)
k = mean_absolute_error(train_predict, train_y)
print('Training Mean Absolute Error', k )
```

```python
test_predict = lr.predict(test_x)
k = mean_absolute_error(test_predict, test_y)
print('Test Mean Absolute Error    ', k )
```
